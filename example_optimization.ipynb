```python
from docplex.mp.model import Model
```

```python
mdl = Model('optimization_problem')
```

### Variables

$x_1 = $ producción de producto tipo 1 <br>
$x_2 = $ producción de producto tipo 2

```python
x1 = mdl.continuous_var(name = 'x1')
x2 = mdl.continuous_var(name = 'x2')
```

### Función objetivo
\begin{equation}
\text{Max } C = 27x_1+21x_2-(10x_1+9x_2)-(14x_1+10x_2)
\end{equation}

```python
mdl.maximize(27*x1+21*x2-(10*x1+9*x2)-(14*x1+10*x2))
```

### Restricciones
\begin{equation}
x_1+x_2\leq 80
\end{equation}
\begin{equation}
2x_1+x_2\leq 100
\end{equation}
\begin{equation}
x_1\leq 40
\end{equation}
\begin{equation}
x_1,x_2 \geq 0
\end{equation}

```python
mdl.add_constraint(x1+x2<=80)
mdl.add_constraint(2*x1+x2<=100)
mdl.add_constraint(x1<=40)
mdl.add_constraint(x1>=0)
mdl.add_constraint(x2>=0)
```

```python
print(mdl.export_to_string())
```

```python
solucion = mdl.solve(log_output = True)
```
